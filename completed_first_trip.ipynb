```python
import pandas as pd # type: ignore
import matplotlib.pyplot as plt # type: ignore
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('ds_challenge_v2_1_data.csv')
df
```

```python
completed_trip = df['first_completed_date'].notnull()
completed_trip_count = completed_trip.sum()
completed_trip_count
```

```python
total_signups = len(df)
labels = ['Completed First Trip', 'Did Not Complete First Trip']
sizes = [completed_trip_count, total_signups - completed_trip_count]
```

```python
colors = ['lightgreen', 'darkgreen']
plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', colors=colors)
plt.title('Fraction of Drivers Who Completed First Trip')
plt.axis('equal') 
plt.show()
```
